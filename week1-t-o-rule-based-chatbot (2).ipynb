```python
#Táº¡o chatbot dáº¡ng rule-based(dá»±a trÃªn tá»« khÃ³a)-táº¡o mennu quÃ¡n cÃ  phÃª
#import thÆ° viá»‡n ipywidgets táº¡o cÃ¡c widget tÆ°Æ¡ng tÃ¡c trong jupyter notebook 
#imprt thÆ° viá»‡n display vÃ  HTML tá»« ipython.display dÃ¹ng hiá»ƒn thá»‹ HTML
import ipywidgets as widgets
from IPython.display import display, HTML
# Tá»« khÃ³a vÃ  pháº£n há»“i tÆ°Æ¡ng á»©ng
rules = {
    "giá» má»Ÿ cá»­a": "QuÃ¡n má»Ÿ cá»­a tá»« 7h sÃ¡ng Ä‘áº¿n 10h tá»‘i.",
    "latte": "Latte cÃ³ giÃ¡ 45.000Ä‘.",
    "cappuccino": "Cappuccino cÃ³ giÃ¡ 50.000Ä‘.",
    "trÃ ": "ChÃºng tÃ´i cÃ³ trÃ  Ä‘Ã o, trÃ  váº£i vÃ  trÃ  xanh.",
    "menu": "Menu gá»“m cÃ  phÃª, trÃ , bÃ¡nh ngá»t vÃ  nÆ°á»›c Ã©p.",
    "wifi": "QuÃ¡n cÃ³ wifi miá»…n phÃ­, máº­t kháº©u lÃ  'caphe123'.",
    "Ä‘á»‹a chá»‰": "Äá»‹a chá»‰ quÃ¡n: 1 vÃµ vÄƒn ngÃ¢n, Thá»§ Ä‘á»©c, TP.HCM.",
    "coffee": "ChÃºng tÃ´i cÃ³ nhiá»u loáº¡i cÃ  phÃª: Ä‘en, sá»¯a, latte, v.v.",
    "cafe": "ChÃºng tÃ´i cÃ³ nhiá»u loáº¡i cÃ  phÃª: Ä‘en, sá»¯a, latte, v.v.",
    "cÃ  phÃª": "ChÃºng tÃ´i cÃ³ nhiá»u loáº¡i cÃ  phÃª: Ä‘en, sá»¯a, latte, v.v.",
}
# Gom nhÃ³m tá»« Ä‘á»“ng nghÄ©a Ä‘á»ƒ chuáº©n hÃ³a tá»« Ä‘á»“ng nghÄ©a vÃ­ dá»¥ ngÆ°á»i dÃ¹ng nháº­p sai chÃ­nh táº£ váº«n hiá»ƒu
synonyms = {
    "cafe": "cÃ  phÃª",
    "coffee": "cÃ  phÃª",
    "cÃ  phÃª": "cÃ  phÃª",
}

#1. HÃ m xá»­ lÃ½ Ä‘áº§u vÃ o chuyá»ƒn Ä‘áº§u vÃ o vá» chá»¯ thÆ°á»ng khÃ´ng phÃ¢n biá»‡t hoa thÆ°á»ng 
#2.thay tá»« Ä‘á»“ng nghÄ©a báº±ng tá»« chuáº©n
#3.so sÃ¡nh vá»›i cÃ¡c tá»« khÃ³a trong rules Ä‘á»ƒ tráº£ lá»i
#4.náº¿u khÃ´ng tÃ¬m tháº¥y tá»« khÃ³a, tráº£ vá» cÃ¢u xin lá»—i

def chatbot_response(user_input):
    user_input = user_input.lower()
    for word in synonyms:
        if word in user_input:
            user_input = user_input.replace(word, synonyms[word])
    for keyword in rules:
        if keyword in user_input:
            return rules[keyword]
    return "Xin lá»—i, tÃ´i chÆ°a hiá»ƒu cÃ¢u há»i cá»§a báº¡n. Vui lÃ²ng thá»­ láº¡i."

# Giao diá»‡n ngÆ°á»i dÃ¹ng Ã´ nháº­p liá»‡u
input_box = widgets.Text(
    placeholder='Nháº­p cÃ¢u há»i cá»§a báº¡n...',
    description='Báº¡n:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='80%')
)
#há»™p xuáº¥t káº¿t quáº£
output_box = widgets.Output()
#hÃ m xá»­ lÃ½ sá»± kiá»‡n khi nháº¥n enter
def observe(sender):
    output_box.clear_output() #xÃ³a káº¿t quáº£ cÅ© trÆ°á»›c khi hiá»ƒn thá»‹ káº¿t quáº£ má»›i
    response = chatbot_response(input_box.value)
    with output_box:
        display(HTML(f"""
        <div style='background-color:#e3f2fd; padding:10px; border-radius:10px;'>
            <div style='color:#0d6efd; font-weight:bold; font-size:16px;'>ğŸ¤– Bot:</div>
            <div style='color:#212529; font-size:15px;'>{response}</div>
        </div>
        """))
#khi ngÆ°á»i dÃ¹ng nháº¥n enter trong Ã´ nháº­p hÃ m observe sáº½ Ä‘Æ°á»£c gá»i
input_box.observe(observe)

# Hiá»ƒn thá»‹ giao diá»‡n vá»›i ná»n xanh dÆ°Æ¡ng
display(HTML("""
<div style='background-color:#0d6efd; padding:20px; border-radius:15px; color:white; text-align:center;'>
    <h2 style='font-size:28px; font-weight:bold;'>ğŸ’¬ Chatbot QuÃ¡n CÃ  PhÃª </h2>
    <p style='font-size:16px;'>Há»i báº¥t cá»© Ä‘iá»u gÃ¬ vá» quÃ¡n cÃ  phÃª cá»§a chÃºng tÃ´i!</p>
</div>
"""))
display(input_box)
display(output_box)
```
