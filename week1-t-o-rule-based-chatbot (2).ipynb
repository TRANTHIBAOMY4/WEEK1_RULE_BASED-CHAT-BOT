```python
#Tạo chatbot dạng rule-based(dựa trên từ khóa)-tạo mennu quán cà phê
#import thư viện ipywidgets tạo các widget tương tác trong jupyter notebook 
#imprt thư viện display và HTML từ ipython.display dùng hiển thị HTML
import ipywidgets as widgets
from IPython.display import display, HTML
# Từ khóa và phản hồi tương ứng
rules = {
    "giờ mở cửa": "Quán mở cửa từ 7h sáng đến 10h tối.",
    "latte": "Latte có giá 45.000đ.",
    "cappuccino": "Cappuccino có giá 50.000đ.",
    "trà": "Chúng tôi có trà đào, trà vải và trà xanh.",
    "menu": "Menu gồm cà phê, trà, bánh ngọt và nước ép.",
    "wifi": "Quán có wifi miễn phí, mật khẩu là 'caphe123'.",
    "địa chỉ": "Địa chỉ quán: 1 võ văn ngân, Thủ đức, TP.HCM.",
    "coffee": "Chúng tôi có nhiều loại cà phê: đen, sữa, latte, v.v.",
    "cafe": "Chúng tôi có nhiều loại cà phê: đen, sữa, latte, v.v.",
    "cà phê": "Chúng tôi có nhiều loại cà phê: đen, sữa, latte, v.v.",
}
# Gom nhóm từ đồng nghĩa để chuẩn hóa từ đồng nghĩa ví dụ người dùng nhập sai chính tả vẫn hiểu
synonyms = {
    "cafe": "cà phê",
    "coffee": "cà phê",
    "cà phê": "cà phê",
}

#1. Hàm xử lý đầu vào chuyển đầu vào về chữ thường không phân biệt hoa thường 
#2.thay từ đồng nghĩa bằng từ chuẩn
#3.so sánh với các từ khóa trong rules để trả lời
#4.nếu không tìm thấy từ khóa, trả về câu xin lỗi

def chatbot_response(user_input):
    user_input = user_input.lower()
    for word in synonyms:
        if word in user_input:
            user_input = user_input.replace(word, synonyms[word])
    for keyword in rules:
        if keyword in user_input:
            return rules[keyword]
    return "Xin lỗi, tôi chưa hiểu câu hỏi của bạn. Vui lòng thử lại."

# Giao diện người dùng ô nhập liệu
input_box = widgets.Text(
    placeholder='Nhập câu hỏi của bạn...',
    description='Bạn:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='80%')
)
#hộp xuất kết quả
output_box = widgets.Output()
#hàm xử lý sự kiện khi nhấn enter
def observe(sender):
    output_box.clear_output() #xóa kết quả cũ trước khi hiển thị kết quả mới
    response = chatbot_response(input_box.value)
    with output_box:
        display(HTML(f"""
        <div style='background-color:#e3f2fd; padding:10px; border-radius:10px;'>
            <div style='color:#0d6efd; font-weight:bold; font-size:16px;'>🤖 Bot:</div>
            <div style='color:#212529; font-size:15px;'>{response}</div>
        </div>
        """))
#khi người dùng nhấn enter trong ô nhập hàm observe sẽ được gọi
input_box.observe(observe)

# Hiển thị giao diện với nền xanh dương
display(HTML("""
<div style='background-color:#0d6efd; padding:20px; border-radius:15px; color:white; text-align:center;'>
    <h2 style='font-size:28px; font-weight:bold;'>💬 Chatbot Quán Cà Phê </h2>
    <p style='font-size:16px;'>Hỏi bất cứ điều gì về quán cà phê của chúng tôi!</p>
</div>
"""))
display(input_box)
display(output_box)
```
